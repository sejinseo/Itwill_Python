성별 팁의 차이가 있을까?

```python
# 남성들의 팁의 평균
tips[tips['sex'] == 'Male'].tip.mean()
```

```python
# 여성들의 팁의 평균
tips[tips['sex'] == 'Female'].tip.mean()
```

```python
# 남성들의 팁의 최댓값
tips[tips['sex'] == 'Male']['tip'].max()
```

```python
# 여성들의 팁의 최댓값
tips[tips['sex'] == 'Female']['tip'].max()
```

```python
# 성별 컬럼의 중복되지 않는 값들
tips['sex'].unique()
```

```python
# 성별 팁의 평균
for s in tips['sex'].unique(): # for s in ['Female', 'Male']:
    subset = tips[tips['sex'] == s] # 성별 부분집합(DataFrame)
    tip_mean = subset['tip'].mean() # 성별 팁의 평균
    print(f'{s} {tip_mean}') # 결과 출력
```

```python
tips.groupby(by='sex')['tip'].mean()
```

흡연 여부별로 팁의 차이가 있을까?

```python
# 흡연 여부별 팁의 평균
tips.groupby(by='smoker')['tip'].mean()
```

요일별 팁의 차이가 있을까?

```python
# 요일별 팁의 평균
tips.groupby(by='day')['tip'].mean()
```

# 성별, 흡연여부별 팁의 차이가 있을까?

```python
tips.groupby(by=['sex','smoker'])['tip'].mean()
```

성별, 요일별 팁의 차이가 있을까?

```python
tips.groupby(by=['sex', 'day'])['tip'].mean()
```

## 여러가지 다양한 groupby

```python
# 성별 total_bill과 tip의 평균
tips.groupby(by='sex')[['total_bill', 'tip']].mean()
```

```python
# 성별, 흡연 여부별 total_bill과 tip의 평균
tips.groupby(by=['sex', 'smoker'])[['total_bill', 'tip']].mean()
```

```python
# 성별 팁의 평균, 최댓값, 최솟값
tips.groupby(by='sex')['tip'].agg(['mean', 'max', 'min']) # aggregate
```

```python
import numpy as np
```

```python
tips.groupby(by='sex')['tip'].agg([np.mean, np.max, np.min])
```

```python
# 성별, 요일별 팁의 평균, 표준편차
tips.groupby(by=['sex', 'day'])['tip'].agg([np.mean, np.std])
```

```python
# 성별, 요일별, 영수증 금액과 팁의 평균, 표준편차
tips.groupby(by=['sex', 'day'])[['total_bill', 'tip']].agg([np.mean, np.std])
```
