* GitHub에 저장된 gapminder.tsv 파일을 읽고 DataFrame을 생성
* 데이터 프레임 탐색
* (중복되지 않는) 국가 이름의 개수?
* (중복되지 않는) 대륙 이름?
* (중복되지 않는) 연도의 개수?
* 국가이름, 연도, 1인당 GDP 컬럼을 선택
* 1인당 GDP의 최댓값, 최솟값을 찾고, 그 row를 출력
* 기대수명(lifeExp)의 최솟값과 최댓값을 찾고, 그 row를 출력
* 연도별 기대수명의 평균을 출력하고 시각화
* 연도별 1인당 GDP의 평균을 출력하고 시각화
* 연도별 대륙별 전세계 인구를 출력하고 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'https://raw.githubusercontent.com/JakeOh/202110_itw_lab_python/main/gapminder.tsv'
df = pd.read_csv(file_path, sep='\t')
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(12, 4))

sns.boxplot(y='lifeExp', data=df, ax=ax[0])
sns.boxplot(y='pop', data=df, ax=ax[1])
sns.boxplot(y='gdpPercap', data=df, ax=ax[2])

plt.show()
```

```python
# 나라 이름 개수
len(df['country'].unique())
```

```python
# 대륙 이름 개수
len(df['continent'].unique())
```

```python
# 연도 개수
len(df['year'].unique())
```

```python
df.columns
```

```python
# fancy indexing - 여러개의 컬럼을 함께 선택하는 방법.
df[['country', 'year', 'gdpPercap']]
```

```python
# 1인당 GDP 최솟값
df['gdpPercap'].min()
```

```python
# 1인당 GDP 최댓값
df['gdpPercap'].max()
```

```python
# 1인당 GDP 최솟값인 행과 최댓값인 행 - boolean indexing(&, |, ~)
df[(df['gdpPercap'] == df['gdpPercap'].min()) |
   (df['gdpPercap'] == df['gdpPercap'].max())]
```

```python
# 기대 수명 최솟값, 최댓값인 행
df[(df['lifeExp'] == df['lifeExp'].min()) |
   (df['lifeExp'] == df['lifeExp'].max())]
```

```python
# Japan의 기대 수명 연도별 그래프
japan = df[df['country'] == 'Japan']
japan
```

```python
sns.lineplot(x='year', y='lifeExp', data=japan, marker='o')
plt.grid()
plt.show()
```

```python
rwanda = df[df['country'] == 'Rwanda']
rwanda
```

```python
sns.lineplot(x='year', y='lifeExp', data=rwanda, marker='o')
plt.grid()
plt.show()
```

```python
# 연도별 기대수명의 평균
df.groupby(by='year')['lifeExp'].mean()
```

```python
sns.lineplot(x='year', y='lifeExp', data=df, marker='o', err_style=None)
plt.grid()
plt.show()
```

```python
sns.barplot(x='year', y='lifeExp', data=df)
plt.grid()
plt.show()
```

```python
# 연도별 1인당 GDP 평균
df.groupby(by='year')['gdpPercap'].mean()
```

```python
sns.lineplot(x='year', y='gdpPercap', data=df, err_style=None, marker='o')
plt.grid()
plt.show()
```

```python
sns.barplot(x='year', y='gdpPercap', data=df)
plt.show()
```

```python
# 연도별 대륙별 전세계 인구
df.groupby(by=['year', 'continent'])['pop'].sum()
```

```python
sns.lineplot(x='year', y='pop', hue='continent', data=df, err_style=None, marker='o', estimator=np.sum)
plt.grid()
plt.show()
```
