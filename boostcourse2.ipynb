```python
# 단계 3: 한글 폰트 설정
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.font_manager as fm

# 마이너스 표시 문제
mpl.rcParams['axes.unicode_minus'] = False
	
# 한글 폰트 설정
path = '/usr/share/fonts/truetype/nanum/NanumGothicBold.ttf'
font_name = fm.FontProperties(fname=path, size=18).get_name()
plt.rc('font', family=font_name)
fm._rebuild()
```

```python
df_raw = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/lab_python/data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20220319195848.csv', encoding='cp949')
df_raw.shape
```

```python
df = df_raw.melt(id_vars=['국가(대륙)별', '상품군별', '판매유형별'], 
                 var_name='기간', value_name='백만원')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
int('2019.4/4'.split('.')[0])
```

```python
df['연도'] = df['기간'].map(lambda x : int(x.split('.')[0]))
df.head()
```

```python
'2019.4/4'.split('.')[1].split('/')[0]
```

```python
df['분기'] = df['기간'].map(lambda x : int(x.split('.')[1].split('/')[0]))
df.head()
```

```python
df['백만원'] = df['백만원'].replace("-", pd.np.nan).astype(float)
df.head()
```

```python
df.info()
```

```python
df = df[(df['국가(대륙)별'] != '합계') & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
df.isnull().sum()
```

```python
df_total = df[df['판매유형별'] == "계"].copy()
df_total.head()
```
