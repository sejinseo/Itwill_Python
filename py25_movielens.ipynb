```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 파일이 저장된 경로
movies_file = 'https://github.com/wesm/pydata-book/raw/2nd-edition/datasets/movielens/movies.dat'
```

* movies.dat 파일을 읽고, DataFrame 생성
* DataFrame 탐색.
* 장르가 'Animation'인 영화는 몇 개?
* 장르가 로맨틱 코미디인 영화는 몇 개?
* 어떤 장르의 영화가 가장 많은가?

```python
movies_colnames = ['movie_id', 'title', 'genres']
movies = pd.read_csv(movies_file, sep='::', engine='python', names=movies_colnames)
```

```python
movies.info()
```

```python
movies.head()
```

```python
movies.tail()
```

```python
# Animation 장르의 영화
animations = movies[movies['genres'].str.contains('Animation')]
animations
```

```python
len(animations)
```

```python
# 로맨틱 코미디 장르 - 장르 컬럼에 'Romance'와 'Comedy'가 포함되어 있는 영화들.
romantic_comedies = movies[(movies['genres'].str.contains('Comedy')) & (movies['genres'].str.contains('Romance'))]
romantic_comedies
```

```python
len(romantic_comedies)
```

```python
genres = [] # 개별 영화 장르를 저장할 빈 리스트
for s in movies['genres']:
    genres.extend(s.split(sep='|'))

len(genres)
```

```python
genre_counts = pd.value_counts(genres)
genre_counts #> Series
```

```python
# 장르 상위 7개 시각화
top7 = genre_counts.head(7)
top7
```

```python
sns.barplot(y=top7.index, x=top7.values)
plt.show()
```

```python
bottom7 = genre_counts.tail(7)
bottom7
```

```python
sns.barplot(x=bottom7.values, y=bottom7.index)
plt.show()
```

```python
genre_counts_df = genre_counts.reset_index()
genre_counts_df.columns = ['genre', 'counts']
genre_counts_df
```

```python
sns.barplot(x='counts', y='genre', data=genre_counts_df)
plt.show()
```

```python
# 연도별 영화 개봉 숫자 시각화
movies
```

```python
# title에서 연도 분리 후 저장

year = []
for y in movies['title']:
    year.append(y[-5:-1])
```

```python
# year 컬럼 생성 후 연도 저장
movies['year'] = year
```

```python
# 출력
movies.head()
```

```python
# 연도별 개봉 영화 오름차순 확인
year_counts = pd.value_counts(year).sort_index()
year_counts
```

```python
# Series -> DataFrame으로 변환 후 컬럼 추가
year_counts_df =year_counts.reset_index()
year_counts_df.columns = ['year', 'counts']
year_counts_df
```

```python
plt.figure(figsize=(20, 10))
sns.lineplot(x='year', y='counts', data=year_counts_df)
plt.grid()
plt.show()
```
